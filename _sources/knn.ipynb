#Tugas 3 **K-NN Algorithm IRIS dataset**

**Algoritma k-Nearest Neighbor** merupakan algoritma supervised learning (pembelajaran terawasi)yang berfungsi untuk melakukan klasifikasi suatu data berdasarkan data pembelajaran (train data sets), yang diambil dari k tetangga terdekatnya (neares neighbors). Dengan k merupakan banyaknya tetangga terdekat.


**Tujuan dari algoritma ini yaitu** untuk mengklasifikasikan obyek baru berdasarkan atribut dan sampel-sampel dari data training

**Algoritma K-Nearest neighbor** menggunakan neighborhood classification sebagai nilai prediksi dari instance yang baru.

Berikut ini adalah cara kerja dari algoritma K-NN yaitu:



*  Tentukan jumlah tetangga yang akan dilakukan perhitungan (k), misalnya kita tentukan tiga (3) tetangga terdekat (k=5).
* Menghitung jarak setiap tetangga terhadap rumah tersebut, kemudian urutkan hasilnya berdasarkan jarak, mulai dari jarak yang terkecil ke jarak yang terbesar.
* Kemudian ambil tiga (3) (k) tetangga yang paling dekat, lalu tentukan kelas dari data baru tersebut sesuai dengan tetangga terdekatnya.



**Keuntungan dan kelemahan K-NN**



  * *Keuntungan*
    - Sangat sederhana implementasinya
    - Kuat dalam hal ruang pencarian, misalnya kelas tidak harus linier dipisahkan
    - Efektif untuk menghitung data dalam skala kecil.
    - Beberapa parameter untuk acuan yaitu jarak metrik dan *k*.




  *   *Kelemahan*

      - Perlu untuk menentukan nilai k yang optimal sehingga untuk menyatakan jumlah tetangga terdekatnya lebih mudah.
      - Biaya komputasi yang cukup tinggi karena perhitungan jarak harus dilakukan pada setiap query instance.







Melakukan load data iris, kemudian membuat variable untuk mendefinisikan data iris

```python
import pandas as pd
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
#the imported dataset does not have the required column names so lets add it
colnames = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'Class']
irisdata = pd.read_csv(url, names=colnames)
```

```python
irisdata
```

Selanjutnya melakukan pendefinisian variable x dan y

```python
#feature matrix in a object named X
X = irisdata.values[:, 0:4]
#response vector in a object named y
y = irisdata.values[:, 4]
```

```python
print(X.shape)
print(y.shape)
```

Mengambil data latih dan data uji

```python
# splitting the data into training and test sets (80:20)
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=4)
```

```python
#shape of train and test objects
print(X_train.shape)
print(X_test.shape)
```

```python
# shape of new y objects
print(y_train.shape)
print(y_test.shape)
```

Melakukan Import library KNeighborsClassifier dari sklearn

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X,y)
```

Malakukan pengecekan accuracy model knn

```python
knn.score(X_test, y_test)
```

Melakukan percobaan untuk mencari spesies yang disesuaikan dengan data baru yang telah diinputkan

```python
#0 = setosa, 1=versicolor, 2=virginica
classes = {0:'Iris-setosa',1:'Iris-versicolor',2:'Iris-virginica'}

#Making prediction on some unseen data 
#predict for the below two random observations
x_new = [[5.8,	4.0,	1.2,	0.2]]
y_predict = knn.predict(x_new)

print([x_new[0],[y_predict[0]]])
```
