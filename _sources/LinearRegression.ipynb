## **Linear Regression**

Regresi Linear digunakan untuk memprediksi suatu pola 

Regresi linear : Alat ukur yang digunakan untuk  mengetahui ada atau tidaknya korelasi antar variabel

Rumus 

Persamaan:  

        Y =aX+b

      Y = variabel terikat

      X = Variabel bebas

      b = intercept

      a = koefesien


```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression
from matplotlib import pyplot
X_test, y_test = make_regression(n_samples=150, n_features=1, noise=0.2)
reg =LinearRegression().fit(X_test, y_test)
reg.score(X_test, y_test)

reg.coef_

reg.intercept_

#reg.predict(np.array[[3]])
```
