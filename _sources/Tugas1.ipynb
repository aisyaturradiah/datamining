**Rumus Mean**

Dalam statistika dan matematika, mean disebut juga dengan rata-rata. Jadi, definisi mean adalah nilai rata-rata dari sekumpulan data yang terdiri dari dua atau lebih..
Berikut rumus mean untuk data tunggal:
$$\bar{x}=\frac{\sum x_i}{n}
$$
Keterangan:

$\bar{x}:$ nilai rata-rata

$\mathrm{x}_{\mathrm{i}}$ : nilai data ke-i

n : banyaknya data

```python
mean = {'Age':np.mean(df['Age']),
		'Total_Bilirubin': np.mean(df['Total_Bilirubin']),'Direct_Bilirubin':np.mean(df['Direct_Bilirubin']),
    	'Alkaline_Phosphotase': np.mean(df['Alkaline_Phosphotase']),'Alamine_Aminotransferase':np.mean(df['Alamine_Aminotransferase']),
     	'Aspartate_Aminotransferase': np.mean(df['Aspartate_Aminotransferase']),'Total_Protiens':np.mean(df['Total_Protiens']),
      	'Albumin': np.mean(df['Albumin']),'Albumin_and_Globulin_Ratio':np.mean(df['Albumin_and_Globulin_Ratio']),
        	'Dataset': np.mean(df['Dataset'])
    }

c = pd.DataFrame(mean, index=['mean'])
c
```

**Rumus Median**

Median disebut juga dengan nilai tengah. Median diperoleh dengan cara mengambil nilai tengah pada data yang telah disusun secara berurutan dari angka terkecil ke terbesar.
Pada data tunggal untuk menentukan median yaitu:
Misalkan banyaknya data n, 

Untuk $n$ ganjil: $M e=x_{\frac{1}{2}(n+1)}$

Untuk $n$ genap: $M e=\frac{\frac{x_n}{2}+\frac{x_n}{2}+1}{2}$

Keterangan: 
$x_{\frac{n}{2}}=$ data pada urutan ke- $\frac{n}{2}$ setelah diurutkan.

```python
median =  {'Age':np.median(df['Age']),
		'Total_Bilirubin': np.median(df['Total_Bilirubin']),'Direct_Bilirubin':np.median(df['Direct_Bilirubin']),
    	'Alkaline_Phosphotase': np.median(df['Alkaline_Phosphotase']),'Alamine_Aminotransferase':np.median(df['Alamine_Aminotransferase']),
     	'Aspartate_Aminotransferase': np.median(df['Aspartate_Aminotransferase']),'Total_Protiens':np.median(df['Total_Protiens']),
      	'Albumin': np.median(df['Albumin']),'Albumin_and_Globulin_Ratio':np.median(df['Albumin_and_Globulin_Ratio']),
        	'Dataset': np.median(df['Dataset'])
    }

c = pd.DataFrame(median, index=['median'])
c
```

**Rumus Standar Deviasi**

Standar deviasi atau simpangan baku biasanya diajarkan pada ilmu statistik untuk mengukur tingkat kesamaan atau kedekatan dalam suatu kelompok.

Standar deviasi adalah nilai statistik yang sering kali dipakai dalam menentukan kedekatan sebaran data yang ada di dalam sampel dan seberapa dekat titik data individu dengan mean atau rata-rata nilai dari sampel itu sendiri.

Untuk Data Tunggal :
$S=\sqrt{\frac{\sum\left(x_i-\bar{x}\right)^2}{n}}$

$S=$ Standar deviasi

$x_i=$ nilai $\mathrm{x} \mathrm {ke-i}$

$\bar{x}=$ nilai rata-rata data

$n=$ jumlah data

```python
import numpy as np
import statistics as st
from scipy import stats
std = {'Age':st.stdev(df['Age']),
		'Total_Bilirubin':st.stdev(df['Total_Bilirubin']),'Direct_Bilirubin':st.stdev(df['Direct_Bilirubin']),
		'Alkaline_Phosphotase':st.stdev(df['Alkaline_Phosphotase']),'Alamine_Aminotransferase':st.stdev(df['Alamine_Aminotransferase']),
		'Aspartate_Aminotransferase':st.stdev(df['Aspartate_Aminotransferase']),'Total_Protiens':st.stdev(df['Total_Protiens']),
		'Albumin':st.stdev(df['Albumin']),'Albumin_and_Globulin_Ratio':st.stdev(df['Albumin_and_Globulin_Ratio']),
		'Dataset':st.stdev(df['Dataset'])
	}

c = pd.DataFrame(std, index=['std'])
c
```

RUMUS PERSENTIL DATA TUNGGAL
$$
P_i=\text { data } k e-\frac{i(n+1)}{100}
$$

Jika persentil 25% maka:
$$
P_i=\text { data } k e-\frac{25(n+1)}{100}
$$

Jika persentil 50% maka:
$$
P_i=\text { data } k e-\frac{50(n+1)}{100}
$$

Jika persentil 75% maka:
$$
P_i=\text { data } k e-\frac{75(n+1)}{100}
$$

Keterangan:

$\mathrm{P}=$ persentil ke $-\mathrm{i}$

$$
i=1,2,3, \ldots, 99
$$
n = banyak data



```python
percentile25 = {'Age':np.percentile(df['Age'],25),
		'Total_Bilirubin':np.percentile(df['Total_Bilirubin'],25),'Direct_Bilirubin':np.percentile(df['Direct_Bilirubin'],25),
  'Alkaline_Phosphotase':np.percentile(df['Alkaline_Phosphotase'],25),'Alamine_Aminotransferase':np.percentile(df['Alamine_Aminotransferase'],25),
  'Aspartate_Aminotransferase':np.percentile(df['Aspartate_Aminotransferase'],25),'Total_Protiens':np.percentile(df['Total_Protiens'],25),
  'Albumin':np.percentile(df['Albumin'],25),'Albumin_and_Globulin_Ratio':np.percentile(df['Albumin_and_Globulin_Ratio'],25),
  'Dataset':np.percentile(df['Dataset'],25)
    }

c = pd.DataFrame(percentile25, index=['25%'])
c
```

## 3.   Mendeskripsikan kemiringan/skewned dari data setiap variabel(bertipe numerik)

*Kemencengan atau kecondongan (skewness)* adalah tingkat ketidaksimetrisan atau kejauhan simetri dari sebuah distribusi. Skewness diartikan sebagai kemiringan distribusi data. Sebuah distribusi yang tidak simetris akan memiliki rata-rata, median, dan modus yang tidak sama besarnya sehingga distribusi akan terkonsentrasi pada salah satu sisi dan kurvanya akan menceng. Ukuran kemiringan kurva adalah derajat atau ukuran dari ketidaksimetrian suatu distribusi data.
